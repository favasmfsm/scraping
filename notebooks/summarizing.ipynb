```python
import pandas as pd
```

```python
import os

csvs = os.listdir("../data/states_data_temp/")
```

```python
df = pd.DataFrame()
for file in csvs:
    df = pd.concat([df,pd.read_csv(f"../data/states_data_temp/{file}")])
```

```python
df
```

```python
df['submission_date'] = pd.to_datetime(df['submission_date'], errors='coerce')

result = df.groupby('state').agg(
    min_date=('submission_date', 'min'),
    max_date=('submission_date', 'max'),
    count=('submission_date', 'count'),
    non_empty_form_name=('form_name', lambda x: (x != '[]').sum())
)
```

```python
result.reset_index(inplace=True)
```

```python
result.to_csv("../outputs/summarized_result_across_states.csv",index=False)
```
