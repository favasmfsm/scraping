```python
import PyPDF2
import textstat

# Extract text
with open("../downloads/SD/127332453/CIGROUPPOL R 1008rev South Dakota.pdf", "rb") as f:
    reader = PyPDF2.PdfReader(f)
    text = " ".join(page.extract_text() for page in reader.pages)

# Compute Flesch score
score = textstat.flesch_reading_ease(text)
print(score)
```

```python
import pandas as pd
df = pd.read_csv("../outputs/temp_results_SD_71608.csv")
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
import ast
df['form_name'] = df.form_name.apply(lambda x: ast.literal_eval(x))
df['flesch_reading_ease'] = df.flesch_reading_ease.apply(lambda x: ast.literal_eval(x))
```

```python
df = df.explode(['form_name', 'flesch_reading_ease'], ignore_index=True)
```

```python
df.to_csv("../outputs/flesch_sample.csv",index=False)
```
