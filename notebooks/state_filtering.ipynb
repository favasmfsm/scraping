```python
import pandas as pd
import os
```

```python
import os
import re

files = os.listdir()
states = set()

for f in files:
    match = re.match(r"temp_results_([A-Z]{2})_", f)
    if match:
        states.add(match.group(1))

print(sorted(states))
```

```python
form_df = pd.read_csv("form_data.csv")
```

```python
df = form_df.copy()
```

```python
df[~df.state.isin(states)].to_csv("to_fetch_states.csv",index=False)
```
